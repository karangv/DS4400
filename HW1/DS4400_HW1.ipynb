#Problem 4.2

```python
# random seed for consistency
np.random.seed(42)

# 3 random 3x3 matrices with integers from -10 to 10
M1 = np.random.randint(-10, 11, size=(3, 3))
M2 = np.random.randint(-10, 11, size=(3, 3))
M3 = np.random.randint(-10, 11, size=(3, 3))
```

```python
# Matrix 1
print("\nM1 =")
print(M1)
det1 = np.linalg.det(M1)
print(f"\nDeterminant of M1: {det1:.4f}")

if np.abs(det1) < 1e-10:
    print("\nMatrix 1 is singular (non-invertible).")
else:
    M1_inv = np.linalg.inv(M1)
    print("\nM1^(-1) =")
    print(M1_inv)
    print("\nM1 × M1^(-1) =")
    print(np.round(M1 @ M1_inv, 10))
```

```python
# Matrix 2
print("\nM2 =")
print(M2)
det2 = np.linalg.det(M2)
print(f"\nDeterminant of M2: {det2:.4f}")

if np.abs(det2) < 1e-10:
    print("\nMatrix 2 is singular (non-invertible).")
else:
    M2_inv = np.linalg.inv(M2)
    print("\nM2^(-1) =")
    print(M2_inv)
    print("\nM2 × M2^(-1) =")
    print(np.round(M2 @ M2_inv, 10))
```

```python
# Matrix 3
print("\nM3 =")
print(M3)
det3 = np.linalg.det(M3)
print(f"\nDeterminant of M3: {det3:.4f}")

if np.abs(det3) < 1e-10:
    print("\nMatrix 3 is singular (non-invertible).")
else:
    M3_inv = np.linalg.inv(M3)
    print("\nM3^(-1) =")
    print(M3_inv)
    print("\nM3 × M3^(-1) =")
    print(np.round(M3 @ M3_inv, 10))
```

#Problem 5

```python
# Load the dataset
df = pd.read_csv('kc_house_data.csv')

# Drop columns to ignore
df = df.drop(columns=['id', 'date', 'zipcode'])

# Separate features and response
response = 'price'
features = [col for col in df.columns if col != response]
```

##5.1

```python
# Compute statistics for each feature
stats = pd.DataFrame({
    'Average': df[features].mean(),
    'Min': df[features].min(),
    'Max': df[features].max(),
    'Variance': df[features].var()
})

print("\n--- Statistics for Each Feature ---\n")
print(stats.to_string())

# Lowest and highest average
lowest_avg_feature = stats['Average'].idxmin()
highest_avg_feature = stats['Average'].idxmax()
print("\n--- Average Analysis ---")
print(f"Feature with LOWEST average:  {lowest_avg_feature} = {stats.loc[lowest_avg_feature, 'Average']:.4f}")
print(f"Feature with HIGHEST average: {highest_avg_feature} = {stats.loc[highest_avg_feature, 'Average']:.4f}")

# Lowest and highest variance
lowest_var_feature = stats['Variance'].idxmin()
highest_var_feature = stats['Variance'].idxmax()
print("\n--- Variance Analysis ---")
print(f"Feature with LOWEST variance:  {lowest_var_feature} = {stats.loc[lowest_var_feature, 'Variance']:.4f}")
print(f"Feature with HIGHEST variance: {highest_var_feature} = {stats.loc[highest_var_feature, 'Variance']:.4f}")
```

##5.2

```python
# Compute correlation of each feature with the response
correlations = df[features].corrwith(df[response])
corr_df = pd.DataFrame({
    'Feature': correlations.index,
    'Correlation with Price': correlations.values
}).sort_values(by='Correlation with Price', ascending=False)

print("\n--- Correlation Coefficients Table ---\n")
print(corr_df.to_string(index=False))

# Positively correlated features
positive_corr = correlations[correlations > 0]
print("\n--- Positively Correlated Features ---")
for feat, corr in positive_corr.sort_values(ascending=False).items():
    print(f"  {feat}: {corr:.4f}")

# Feature with highest positive correlation
highest_corr_feature = correlations.idxmax()
highest_corr_value = correlations.max()
print(f"\nFeature with HIGHEST positive correlation: {highest_corr_feature} = {highest_corr_value:.4f}")
```

##5.3

```python
negative_corr = correlations[correlations < 0]
if len(negative_corr) > 0:
    print("\n--- Features with Negative Correlation ---")
    for feat, corr in negative_corr.sort_values().items():
        print(f"  {feat}: {corr:.4f}")
else:
    print("\nNo features with negative correlation coefficient were found.")
```
